```python
import pandas as pd
```

```python
df = pd.read_csv('../images/results.csv')
print(df)
```

```python
def build_frame(cnt):
    return df[df['triangles'] == cnt]

tris = [94012, 618490]
tri_data = list(map(build_frame, tris))
```

```python
def plot_data(data, tri_count):
    data.plot.bar(x = 'accelerator', y = 'build', rot = 0, title = str(tri_count) + ' triangles')
    data.plot.bar(x = 'accelerator', y = 'render', rot = 0, title = str(tri_count) + ' triangles')
    
for index, data in enumerate(tri_data):
    plot_data(data, tris[index])
```

```python
def build_frame(name):
    return df[df['accelerator'] == name]

accels = ['bvh', 'kd_tree', 'madman_bvh', 'fast_bvh']
accel_data = list(map(build_frame, accels))
```

```python
def plot_data(data, tri_count):
    data.plot.bar(x = 'triangles', y = 'build', rot = 0, title = str(tri_count) + ' triangles')
    data.plot.bar(x = 'triangles', y = 'render', rot = 0, title = str(tri_count) + ' triangles')
    
for index, data in enumerate(accel_data):
    plot_data(data, accels[index])
```
